```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os
```

```python
# Read in the file
game_data_csv = "purchasedata.csv"
gamedata_df = pd.read_csv(game_data_csv)
gamedata_df.head()
```

```python
# Total Players Count
totalplayers_count = len(gamedata_df["SN"].unique())
totalplayers_count

# Create new df
totalplayers_df = pd.DataFrame({"Total Players": [totalplayers_count]})

totalplayers_df
```

```python
# Purchase Analysis
uniqueitems_count = len(gamedata_df["Item ID"].unique())
uniqueitems_count

averageprice = gamedata_df["Price"].mean()
averageprice

totalpurchases = gamedata_df["Item Name"].count()
totalpurchases

totalrev = gamedata_df["Price"].sum()
totalrev

#New df
totalsanalysis_df = pd.DataFrame({"Total Unique Items": [uniqueitems_count], 
                                  "Average Price": [averageprice],
                                    "Total Purhcases": [totalpurchases],
                                 "Total Revenue": [totalrev]})

totalsanalysis_df["Average Price"] = totalsanalysis_df["Average Price"].map("${:.2f}".format)
totalsanalysis_df["Total Revenue"] = totalsanalysis_df["Total Revenue"].map("${:.2f}".format)


totalsanalysis_df
```

```python
# Gender Demographics
genderinfo = gamedata_df.groupby("Gender")
gendercount = genderinfo.nunique()["SN"]

percentplayers = gendercount / totalplayers_count * 100

genderdemo = pd.DataFrame({"Percentage of Players": percentplayers, "Total Count": gendercount})

genderdemo.index.name = None

genderdemo["Percentage of Players"] = genderdemo["Percentage of Players"].astype(float).map("{:,.2f}%".format)

genderdemo.sort_values(by=["Total Count"], ascending = True)

genderdemo
```

```python
#Purchase Analysis by Gender
purchasescount = genderinfo["Purchase ID"].count()

avgprice = genderinfo["Price"].mean()
avgtotal = genderinfo["Price"].sum()
avgpurchasebyperson = avgtotal / gendercount

# Create new df
purchaseupdated = pd.DataFrame({"Purchase Count": purchasescount, 
                                    "Average Purchase Price": avgprice,
                                    "Average Purchase Value": avgtotal,
                                    "Avg Total Purchase per Person": avgpurchasebyperson
                                   })

#Clean up data
purchaseupdated.index.name = "Gender"

purchaseupdated["Average Purchase Price"] = purchaseupdated["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchaseupdated["Average Purchase Value"] = purchaseupdated["Average Purchase Value"].astype(float).map("${:,.2f}".format)
purchaseupdated["Avg Total Purchase per Person"] = purchaseupdated["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

purchaseupdated
```

```python
# Create Bins and Labels
agesbins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

gamedata_df["Age Group"] = pd.cut(gamedata_df["Age"],agesbins, labels=groups)
gamedata_df

# Calculations
agesgrouped = gamedata_df.groupby("Age Group")
totalages = agesgrouped["SN"].nunique()
agepercentage = (totalages/totalplayers_count) * 100

# Create df
agedemo = pd.DataFrame({"Total Count": totalages, "Percentage of Players": agepercentage})
agedemo.index.name = None

agedemo["Percentage of Players"] = agedemo["Percentage of Players"].astype(float).map("{:,.2f}%".format)
agedemo
```

```python
# Purchasing Analysis by Age
purchases_byage = agesgrouped["Purchase ID"].count()

avgpur_byage = agesgrouped["Price"].mean()
totalpurchase = agesgrouped["Price"].sum()
avgpur_byage = totalpurchase/totalages

# Create df
spendingbyage = pd.DataFrame({"Purchase Count": purchases_byage,
                                 "Average Purchase Price": avgpur_byage,
                                 "Total Purchase Value":totalpurchase,
                                 "Average Purchase Total per Person": avgpur_byage})

spendingbyage.index.name = None

spendingbyage["Average Purchase Price"] = spendingbyage["Average Purchase Price"].astype(float).map("${:,.2f}".format)
spendingbyage["Total Purchase Value"] = spendingbyage["Total Purchase Value"].astype(float).map("${:,.2f}".format)
spendingbyage["Average Purchase Total per Person"] = spendingbyage["Average Purchase Total per Person"].astype(float).map("${:,.2f}".format)

spendingbyage
```

```python
# Top Spenders
topspenders = gamedata_df.groupby("SN")

purchase_totals = topspenders["Purchase ID"].count()
avgpurchase_byspender = topspenders["Price"].mean()
totalpurchase_byspender = topspenders["Price"].sum()

# Create new df
top_spenders = pd.DataFrame({"Purchase Count": purchase_totals,
                             "Average Purchase Price": avgpurchase_byspender,
                             "Total Purchase Value":totalpurchase_byspender})

topspenders_clean = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()

topspenders_clean.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})
```

```python
# Most Popular Items
items = gamedata_df[["Item ID", "Item Name", "Price"]]

# Calculate for item data
iteminfo = items.groupby(["Item ID","Item Name"])
itemcount = iteminfo["Price"].count()
totalpurchasevalue = (iteminfo["Price"].sum()) 
itemprice = totalpurchasevalue/itemcount

# Create new DF
popularitems = pd.DataFrame({"Purchase Count": itemcount, 
                                   "Item Price": itemprice,
                                   "Total Purchase Value":totalpurchasevalue})

popularitems_clean = popularitems.sort_values(["Purchase Count"], ascending=False).head()

popularitems_clean.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```

```python
# Most Profitable Items
popularitems_clean = popularitems.sort_values(["Total Purchase Value"],
                                                   ascending=False).head()

popularitems_clean.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```
